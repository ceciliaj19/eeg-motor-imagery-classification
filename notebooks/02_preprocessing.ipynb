# EEG Preprocessing

## Goal

Prepare raw EEG recordings for motor imagery classification by extracting event markers, filtering the signal, and creating epochs.

```python
import mne
import matplotlib.pyplot as plt

raw = mne.io.read_raw_edf("../data/S001R06.edf", preload=True)

print(raw)
```

```python
events, event_id = mne.events_from_annotations(raw)

print(event_id)
print(events[:10])
```

```python
event_id_clean = {
    "class_1": 2,
    "class_2": 3
}

event_id_clean
```

```python
raw_filtered = raw.copy().filter(8, 30)
```

```python
epochs = mne.Epochs(
    raw_filtered,
    events,
    event_id=event_id_clean,
    tmin=0,
    tmax=4,
    baseline=None,
    preload=True
)

print(epochs)
```

```python
print(epochs)
print(epochs.event_id)
print(epochs.get_data().shape)
```

```python
import mne

files = [
    "../data/S001R06.edf",
    "../data/S001R10.edf",
    "../data/S001R14.edf",
]

raws = [mne.io.read_raw_edf(f, preload=True) for f in files]
raw = mne.concatenate_raws(raws)

print(raw)
```

```python
events, event_id = mne.events_from_annotations(raw)

print(event_id)
print(events.shape)
print(events[:10])
```

```python
raw_filtered = raw.copy().filter(8, 30)
```

```python
event_id_clean = {
    "class_1": 2,
    "class_2": 3
}

epochs = mne.Epochs(
    raw_filtered,
    events,
    event_id=event_id_clean,
    tmin=0,
    tmax=4,
    baseline=None,
    preload=True
)

print(epochs)
print(epochs.get_data().shape)
```

```python
epochs.save("../data/subject1_motor_imagery-epo.fif", overwrite=True)
```

## Preprocessing Result

After combining runs 6, 10, and 14 for Subject 1, I extracted 45 motor imagery epochs. Each epoch contains 64 EEG channels and 641 time points. The data was filtered between 8 and 30 Hz, which has the mu and beta rhythms which are most commonly used in motor imagery BCI classification.
